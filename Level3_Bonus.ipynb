Level 3 Bonus: Create a chart to tell me something about this data: https://github.com/gumdropsteve/datasets/raw/master/airlines.parquet (how to read in below)

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_parquet('https://github.com/gumdropsteve/datasets/raw/master/airlines.parquet')
df_fig = df[['Origin','Dest','ActualElapsedTime']].head()

df_fig.plot.barh(x='ActualElapsedTime',color =['maroon','coral'], title = "The Actual Elapsed Time for each Flight ",fontsize=15 ,  legend = True)
plt.show()
```

## I calculated the actual elapsed time for each trip by calculating the difference between the Origin and the Dest
